```
# THROUGH THE LENS OF TRUTH: FAKE NEWS DETECTION SYSTEM
# TruthLens Institute - Comprehensive Solution
#
# This comprehensive solution includes:
# 1. Data loading and cleaning
# 2. Text preprocessing and normalization
# 3. Feature engineering
# 4. Model training with class imbalance handling
# 5. Model evaluation and interpretation
# 6. Insights generation and recommendations
#
# The system processes news articles, extracts linguistic features,
# and classifies them as real or fake news using a Random Forest classifier.


# =====================
# PHASE 0: SETUP & CONFIGURATION
# =====================
# Import necessary libraries
import pandas as pd  # Data manipulation and analysis
import numpy as np   # Numerical operations
import matplotlib.pyplot as plt  # Data visualization
import seaborn as sns  # Enhanced data visualization
import re  # Regular expressions for text processing
import os  # Operating system interactions
from bs4 import BeautifulSoup  # HTML parsing
import nltk  # Natural Language Toolkit
from nltk.corpus import stopwords  # Common stopwords
from wordcloud import WordCloud  # Word cloud visualization
from textblob import TextBlob  # Sentiment analysis
from sklearn.feature_extraction.text import TfidfVectorizer  # TF-IDF vectorization
from sklearn.model_selection import train_test_split  # Data splitting
from sklearn.ensemble import RandomForestClassifier  # Classification model
from sklearn.metrics import classification_report, f1_score, roc_auc_score, confusion_matrix  # Evaluation metrics
from sklearn.metrics import precision_recall_fscore_support, roc_curve  # Additional metrics
import joblib  # Model persistence
import spacy  # Advanced NLP processing
import time  # Timing operations
from datetime import datetime  # Date/time handling
from imblearn.over_sampling import SMOTE  # Handling class imbalance
import dateutil.parser  # Flexible date parsing
from IPython.display import display, HTML  # Enhanced display in Jupyter

# Configuration settings
pd.set_option('display.max_columns', 50)  # Show up to 50 columns
pd.set_option('display.max_colwidth', 200)  # Show up to 200 characters per column
plt.style.use('ggplot')  # Use ggplot style for plots

# Download NLTK resources
nltk.download(['stopwords', 'wordnet', 'omw-1.4'], quiet=True)

# =====================================================================
# TEXT PROCESSING CONFIGURATION
# =====================================================================
# Attempt to load spaCy for advanced NLP, fallback to NLTK if unavailable
try:
    # Load English model, disabling parser and NER for efficiency
    nlp = spacy.load('en_core_web_sm', disable=['parser', 'ner'])
    USE_SPACY = True
    print("Using spaCy for lemmatization")
except ImportError:
    print("spaCy model not available, using NLTK for lemmatization")
    from nltk.stem import WordNetLemmatizer
    lemmatizer = WordNetLemmatizer()
    USE_SPACY = False

# Set random seeds for reproducibility
RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)

# Create output directory for saving results
os.makedirs('output', exist_ok=True)

# Display helper functions for Jupyter
def display_header(text, level=2):
    """Display formatted header text"""
    display(HTML(f"<h{level} style='color:#1f77b4;'>{text}</h{level}>"))

def display_success(text):
    """Display success messages in green"""
    display(HTML(f"<div style='color:green; font-weight:bold;'>{text}</div>"))

def display_warning(text):
    """Display warning messages in orange"""
    display(HTML(f"<div style='color:orange; font-weight:bold;'>{text}</div>"))

def display_error(text):
    """Display error messages in red"""
    display(HTML(f"<div style='color:red; font-weight:bold;'>{text}</div>"))



# =====================
# PHASE 1: DATA LOADING & CLEANING
# =====================
# This phase focuses on:
# 1. Loading the dataset from CSV
# 2. Handling data quality issues
# 3. Standardizing column names
# 4. Cleaning and preprocessing text content
# 5. Parsing and normalizing dates
# 6. Performing exploratory data analysis

display_header("PHASE 1: DATA LOADING & CLEANING", level=2)

def load_news_data(filepath):
    """
    Robust CSV loader that handles various formatting issues
    - Attempts standard pandas loading first
    - Falls back to manual parsing if needed
    - Handles encoding errors and irregular rows
    """
    display(f"Loading dataset from {filepath}...")
    start_time = time.time()

    # Verify file existence
    if not os.path.exists(filepath):
        display_error(f"Dataset file {filepath} not found")
        raise FileNotFoundError(f"Dataset file {filepath} not found")

    try:
        # Attempt standard CSV loading
        df = pd.read_csv(filepath, dtype=str, on_bad_lines='warn',
                         encoding_errors='replace', low_memory=False)
        display_success(f"Loaded {len(df)} records in {time.time()-start_time:.2f}s")
        return df
    except Exception as e:
        display_warning(f"Standard loading failed: {str(e)}")
        display_warning("Manual CSV parsing not implemented in this version")
        return None

# Load dataset
df = load_news_data('fake.csv')

# Only proceed if data loaded successfully
if df is not None and not df.empty:
    display(f"Total records: {len(df)}")

    # =================================================================
    # DATA QUALITY ASSESSMENT
    # =================================================================
    display_header("Data Quality Report", level=3)

    # Show initial columns
    display("Initial columns:")
    display(df.columns.tolist())

    # Count missing values
    display("\nMissing values per column:")
    display(df.isnull().sum().to_frame('Missing Count'))

    # Show sample data
    display("\nSample data:")
    display(df.head(2))

    # =================================================================
    # COLUMN STANDARDIZATION
    # =================================================================
    # Define critical columns we need for analysis
    CRITICAL_COLS = ['text', 'type', 'site_url', 'published']

    # Map possible column names to standard names
    COLUMN_MAPPING = {
        'text': ['content', 'article', 'body', 'text'],
        'type': ['label', 'category', 'classification', 'type'],
        'site_url': ['link', 'source_url', 'url', 'site_url'],
        'published': ['date', 'timestamp', 'publish_date', 'published']
    }

    def map_columns(df, mapping):
        """Standardize column names across datasets"""
        for standard_name, alternatives in mapping.items():
            for alt in alternatives:
                if alt in df.columns:
                    if standard_name != alt:
                        df[standard_name] = df[alt]
                        display(f"Mapped column '{alt}' to '{standard_name}'")
                    break
        return df

    # Apply column standardization
    df = map_columns(df, COLUMN_MAPPING)

    # Create placeholders for missing critical columns
    for col in CRITICAL_COLS:
        if col not in df.columns:
            display_warning(f"Critical column '{col}' not found - creating placeholder")
            df[col] = ''

    # =================================================================
    # TEXT CLEANING PIPELINE
    # =================================================================
    def clean_text(text):
        """
        Comprehensive text cleaner:
        1. Handles non-string values
        2. Removes HTML tags
        3. Removes URLs
        4. Removes special characters
        5. Normalizes whitespace
        """
        if not isinstance(text, str):
            return ""

        # Parse and remove HTML
        text = BeautifulSoup(text, 'html.parser').get_text()

        # Remove URLs
        text = re.sub(r'https?://\S+|www\.\S+', '', text)

        # Keep only alphanumeric and basic punctuation
        text = re.sub(r'[^\w\s.,!?\'"-]', '', text)

        # Collapse multiple whitespaces
        text = re.sub(r'\s+', ' ', text).strip()

        return text

    display("Cleaning text data...")
    df['clean_text'] = df['text'].apply(clean_text)

    # Remove empty texts
    empty_texts = df['clean_text'].str.len() == 0
    display(f"Found {empty_texts.sum()} empty texts after cleaning - removing them")
    df = df[~empty_texts].reset_index(drop=True)

    # =================================================================
    # TEXT NORMALIZATION
    # =================================================================
    def normalize_text(text):
        """
        Normalize text for consistency:
        1. Convert to lowercase
        2. Remove numbers
        3. Remove special punctuation
        """
        text = text.lower()
        text = re.sub(r'\d+', '', text)  # Remove digits
        text = re.sub(r'[^\w\s\'-]', '', text)  # Keep only word chars and basic punctuation
        return text

    display("Normalizing text...")
    df['normalized_text'] = df['clean_text'].apply(normalize_text)

    # =================================================================
    # TEXT LEMMATIZATION
    # =================================================================
    def lemmatize_text(text):
        """
        Reduce words to their base/dictionary form:
        - Uses spaCy if available for more accurate lemmatization
        - Falls back to NLTK WordNet lemmatizer
        """
        if USE_SPACY:
            doc = nlp(text)
            return " ".join([token.lemma_ for token in doc])
        else:
            words = nltk.word_tokenize(text)
            return " ".join([lemmatizer.lemmatize(word) for word in words])

    display("Lemmatizing text...")
    df['lemmatized_text'] = df['normalized_text'].apply(lemmatize_text)

    # =================================================================
    # DATE PROCESSING
    # =================================================================
    def convert_to_datetime_enhanced(series):
        """
        Robust datetime parser with timezone normalization:
        1. Handles various date formats
        2. Converts timezone-aware dates to UTC
        3. Returns timezone-naive datetime objects
        4. Handles parsing errors gracefully
        """
        parsed_dates = []
        display("Parsing dates with timezone normalization...")

        for date_str in series.astype(str):
            try:
                # Parse with flexible dateutil parser
                dt = dateutil.parser.parse(date_str, fuzzy=True)

                # Convert to UTC and remove timezone info
                if dt.tzinfo is not None:
                    dt = dt.astimezone(dateutil.tz.UTC).replace(tzinfo=None)

                parsed_dates.append(dt)
            except (ValueError, TypeError, dateutil.parser.ParserError):
                parsed_dates.append(pd.NaT)  # Mark as missing on error

        # Convert to datetime series
        result = pd.Series(parsed_dates, dtype='datetime64[ns]')
        valid_count = result.notna().sum()
        display(f"Successfully parsed {valid_count}/{len(result)} dates")
        return result

    # Process publication dates
    if 'published' in df.columns:
        df['publish_date'] = convert_to_datetime_enhanced(df['published'])

        # Fill missing dates with current time
        current_time = pd.Timestamp.now()
        df['publish_date'] = df['publish_date'].fillna(current_time)

        # Extract temporal features
        df['publish_hour'] = df['publish_date'].dt.hour
        df['publish_day'] = df['publish_date'].dt.day_name()
    else:
        # Create placeholder dates if column missing
        current_time = pd.Timestamp.now()
        df['publish_date'] = current_time
        df['publish_hour'] = 0
        df['publish_day'] = 'Unknown'
        display_warning("Published date column not found - using placeholders")

    # =================================================================
    # EXPLORATORY DATA ANALYSIS (EDA)
    # =================================================================
    display_header("Exploratory Data Analysis", level=3)

    # News type distribution
    if 'type' in df.columns and df['type'].nunique() > 1:
        plt.figure(figsize=(10, 6))
        type_counts = df['type'].value_counts()

        # Handle datasets with many categories
        if len(type_counts) > 10:
            # Show top 10 types
            top_types = type_counts.head(10)
            plt.barh(top_types.index.astype(str), top_types.values)
            plt.title('Top 10 News Types')
        else:
            # Show all types
            sns.countplot(y='type', data=df, order=type_counts.index)
            plt.title('Distribution of News Types')
        plt.tight_layout()
        plt.show()
    else:
        display_warning("Insufficient type data for distribution plot")

    # =================================================================
    # WORD CLOUD VISUALIZATION
    # =================================================================
    def generate_word_cloud(text, title):
        """
        Generate word cloud visualization:
        - Filters out stopwords
        - Shows most frequent words
        - Handles empty text inputs
        """
        if not text.strip():
            display_warning(f"Skipping empty word cloud: {title}")
            return

        try:
            # Create and display word cloud
            wordcloud = WordCloud(
                width=800,
                height=400,
                background_color='white',
                stopwords=set(stopwords.words('english')),
                max_words=200
            ).generate(text)

            plt.figure(figsize=(12, 6))
            plt.imshow(wordcloud, interpolation='bilinear')
            plt.axis('off')
            plt.title(title, fontsize=16)
            plt.tight_layout()
            plt.show()
        except Exception as e:
            display_error(f"Word cloud error: {str(e)}")

    # Generate word clouds for news types
    if 'type' in df.columns and df['type'].nunique() > 1:
        # Limit to first 3 types for practical reasons
        for news_type in df['type'].unique()[:3]:
            if news_type and pd.notna(news_type):
                # Sample text for efficiency
                # Adjust sample size based on available data
                type_df = df[df['type'] == news_type]
                current_sample_size = min(500, len(type_df))

                if current_sample_size > 0:
                    type_text = ' '.join(
                        type_df['lemmatized_text']
                        .sample(current_sample_size, random_state=RANDOM_STATE)
                    )
                    display_header(f"Word Cloud: {news_type}", level=4)
                    generate_word_cloud(type_text, f'Most Frequent Words: {news_type}')
                else:
                    display_warning(f"Skipping word cloud for '{news_type}' due to insufficient data.")


    # Save cleaned data
    df.to_csv('output/cleaned_news.csv', index=False)
    display_success("Saved cleaned data to output/cleaned_news.csv")
else:
    display_error("Data loading failed - cannot proceed to next phases")
    df = pd.DataFrame()  # Create empty DataFrame for downstream compatibility



# =====================
# PHASE 2: FEATURE ENGINEERING
# =====================
# This phase focuses on:
# 1. Extracting linguistic features from text
# 2. Calculating sentiment polarity
# 3. Extracting domain information from URLs
# 4. Creating TF-IDF vector representations
# 5. Combining all features into a single matrix

if not df.empty:
    display_header("PHASE 2: FEATURE ENGINEERING", level=2)

    # =================================================================
    # LINGUISTIC FEATURE EXTRACTION
    # =================================================================
    # Word count: Number of words in article
    df['word_count'] = df['lemmatized_text'].apply(lambda x: len(x.split()))

    # Character count: Total characters in article
    df['char_count'] = df['lemmatized_text'].apply(len)

    # Average word length: Character count divided by word count
    df['avg_word_length'] = df['char_count'] / (df['word_count'].replace(0, 1))

    # =================================================================
    # SENTIMENT ANALYSIS
    # =================================================================
    display("Calculating sentiment polarity...")
    df['sentiment'] = df['lemmatized_text'].apply(
        lambda x: TextBlob(x).sentiment.polarity
    )

    # =================================================================
    # DOMAIN EXTRACTION
    # =================================================================
    display("Extracting domain features from URLs...")
    if 'site_url' in df.columns:
        df['domain'] = df['site_url'].apply(
            lambda x: (
                x.split('//')[1].split('/')[0].replace('www.', '')
                if isinstance(x, str) and '//' in x
                else 'unknown'
            )
        )
    else:
        df['domain'] = 'unknown'
        display_warning("site_url column missing - using placeholder domain")

    # =================================================================
    # TF-IDF VECTORIZATION
    # =================================================================
    display("Computing TF-IDF features...")

    # Configure TF-IDF vectorizer
    tfidf = TfidfVectorizer(
        max_features=1000,  # Limit to top 1000 terms for efficiency
        ngram_range=(1, 2),  # Include unigrams and bigrams
        stop_words='english',  # Filter English stopwords
        min_df=5,  # Ignore terms appearing in fewer than 5 documents
        max_df=0.7  # Ignore terms appearing in more than 70% of documents
    )

    # Transform text to TF-IDF features
    tfidf_features = tfidf.fit_transform(df['lemmatized_text'])

    # Convert to DataFrame
    tfidf_df = pd.DataFrame(
        tfidf_features.toarray(),
        columns=tfidf.get_feature_names_out(),
        index=df.index
    )

    # =================================================================
    # FEATURE COMBINATION
    # =================================================================
    # Select metadata features
    feature_cols = ['word_count', 'char_count', 'avg_word_length', 'sentiment', 'publish_hour']

    # Combine metadata with TF-IDF features
    feature_df = pd.concat([df[feature_cols], tfidf_df], axis=1)

    # Fill any remaining missing values
    feature_df = feature_df.fillna(0)
    display(f"Final feature matrix shape: {feature_df.shape}")

    # =================================================================
    # FEATURE DISTRIBUTION VISUALIZATION
    # =================================================================
    plt.figure(figsize=(15, 10))

    # Word count distribution
    plt.subplot(2, 2, 1)
    sns.histplot(df['word_count'], bins=50, kde=True)
    plt.title('Word Count Distribution')

    # Character count distribution
    plt.subplot(2, 2, 2)
    sns.histplot(df['char_count'], bins=50, kde=True)
    plt.title('Character Count Distribution')

    # Average word length distribution
    plt.subplot(2, 2, 3)
    sns.histplot(df['avg_word_length'], bins=50, kde=True)
    plt.title('Average Word Length Distribution')

    # Sentiment distribution
    plt.subplot(2, 2, 4)
    sns.histplot(df['sentiment'], bins=50, kde=True)
    plt.title('Sentiment Polarity Distribution')

    plt.tight_layout()
    plt.show()



# =====================
# PHASE 3: MODEL DEVELOPMENT
# =====================
# This phase focuses on:
# 1. Preparing the target variable (real vs fake)
# 2. Splitting data into training and test sets
# 3. Handling class imbalance with SMOTE
# 4. Training a Random Forest classifier
# 5. Evaluating model performance
# 6. Saving the trained model

if not df.empty and 'feature_df' in locals():
    display_header("PHASE 3: MODEL DEVELOPMENT", level=2)

    # =================================================================
    # TARGET VARIABLE PREPARATION
    # =================================================================
    # Create binary target: 1 for fake news, 0 for real news
    if 'type' in df.columns and df['type'].nunique() > 1:
        # Define keywords that indicate fake news
        fake_labels = [
            'fake', 'false', 'hoax', 'conspiracy', 'satire',
            'junk', 'rumor', 'bs', 'clickbait', 'propaganda'
        ]

        # Create regex pattern for matching
        pattern = '|'.join(fake_labels)

        # Create target series (1 for fake, 0 otherwise)
        y = df['type'].str.lower().str.contains(pattern).astype(int)

        # Check if we have fake news samples
        if y.sum() == 0:
            display_warning("No fake samples detected! Using synthetic labels")
            # Create synthetic labels (80% real, 20% fake)
            y = pd.Series(np.random.choice([0, 1], size=len(df), p=[0.8, 0.2]))
        else:
            display(f"Detected {y.sum()} fake samples ({y.mean():.2%})")
    else:
        display_warning("'type' column missing/invalid - creating synthetic labels")
        # Create synthetic labels (70% real, 30% fake)
        y = pd.Series(np.random.choice([0, 1], size=len(df), p=[0.7, 0.3]))

    # Display target distribution
    display("Target distribution:")
    display(y.value_counts().to_frame('Count'))

    # =================================================================
    # DATA SPLITTING
    # =================================================================
    # Split into training (80%) and test (20%) sets
    # Stratify by target to maintain class distribution
    X_train, X_test, y_train, y_test = train_test_split(
        feature_df, y,
        test_size=0.2,
        random_state=RANDOM_STATE,
        stratify=y
    )

    display(f"Training samples: {X_train.shape[0]}")
    display(f"Test samples: {X_test.shape[0]}")

    # Display class distributions
    train_dist = y_train.value_counts(normalize=True).mul(100).rename('Train %')
    test_dist = y_test.value_counts(normalize=True).mul(100).rename('Test %')
    display("Class distribution:")
    display(pd.concat([train_dist, test_dist], axis=1))

    # =================================================================
    # CLASS IMBALANCE HANDLING
    # =================================================================
    # Count minority class (fake news) samples
    minority_count = sum(y_train == 1)
    display(f"Minority class (fake) samples in training: {minority_count}")

    # Apply SMOTE if sufficient minority samples exist
    if minority_count >= 5:
        display("Applying SMOTE for class balancing...")

        # Configure SMOTE (Synthetic Minority Over-sampling Technique)
        smote = SMOTE(
            sampling_strategy='auto',
            k_neighbors=min(5, minority_count - 1),  # Ensure valid neighbors
            random_state=RANDOM_STATE
        )

        # Resample training data
        X_train, y_train = smote.fit_resample(X_train, y_train)
        display(f"Resampled training samples: {len(X_train)}")

        # Show new distribution
        new_dist = pd.Series(y_train).value_counts(normalize=True).mul(100)
        display("New class distribution:")
        display(new_dist)
    else:
        display_warning("Insufficient minority samples - using class weights")

    # =================================================================
    # MODEL TRAINING
    # =================================================================
    # Configure Random Forest classifier
    model = RandomForestClassifier(
        n_estimators=200,  # Number of trees
        max_depth=15,  # Maximum tree depth
        min_samples_leaf=2,  # Minimum samples per leaf
        class_weight='balanced_subsample' if minority_count < 5 else None,
        random_state=RANDOM_STATE,
        n_jobs=-1  # Use all available cores
    )

    display("Training Random Forest model...")
    start_time = time.time()
    model.fit(X_train, y_train)
    display(f"Training completed in {time.time()-start_time:.2f} seconds")

    # =================================================================
    # MODEL EVALUATION
    # =================================================================
    display_header("MODEL EVALUATION", level=3)

    # Generate predictions
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1]  # Probability of fake class

    # =================================================================
    # CLASSIFICATION REPORT
    # =================================================================
    display("Classification Report:")
    report = classification_report(y_test, y_pred, output_dict=True)
    display(pd.DataFrame(report).transpose())

    # =================================================================
    # KEY METRICS
    # =================================================================
    precision, recall, f1, _ = precision_recall_fscore_support(
        y_test, y_pred, average='binary'
    )
    roc_auc = roc_auc_score(y_test, y_proba)

    # Create metrics summary
    metrics_df = pd.DataFrame({
        'Metric': ['F1 Score', 'ROC-AUC', 'Precision', 'Recall'],
        'Value': [f1, roc_auc, precision, recall]
    })
    display("Key Performance Metrics:")
    display(metrics_df)

    # =================================================================
    # CONFUSION MATRIX
    # =================================================================
    plt.figure(figsize=(8, 6))
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['Real', 'Fake'],
                yticklabels=['Real', 'Fake'])
    plt.title(f'Confusion Matrix (F1={f1:.3f})')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.tight_layout()
    plt.show()

    # =================================================================
    # ROC CURVE
    # =================================================================
    fpr, tpr, thresholds = roc_curve(y_test, y_proba)
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='darkorange', lw=2,
             label=f'ROC curve (AUC = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic')
    plt.legend(loc="lower right")
    plt.show()

    # =================================================================
    # FEATURE IMPORTANCE
    # =================================================================
    if len(feature_df.columns) > 5:
        try:
            plt.figure(figsize=(12, 8))
            importances = model.feature_importances_

            # Get top 20 important features
            indices = np.argsort(importances)[-20:]
            features = feature_df.columns[indices]

            # Create horizontal bar plot
            plt.barh(range(len(indices)), importances[indices], align='center')
            plt.yticks(range(len(indices)), features)
            plt.title('Top 20 Important Features')
            plt.xlabel('Feature Importance Score')
            plt.tight_layout()
            plt.show()
        except Exception as e:
            display_error(f"Feature importance error: {str(e)}")

    # =================================================================
    # MODEL SAVING
    # =================================================================
    joblib.dump(model, 'output/fake_news_detector.pkl')
    display_success("Model saved as output/fake_news_detector.pkl")
else:
    display_error("Feature engineering failed - cannot proceed to modeling")



# =====================
# PHASE 4: INSIGHTS & RECOMMENDATIONS
# =====================
# This phase focuses on:
# 1. Extracting key insights from the analysis
# 2. Visualizing patterns in the data
# 3. Discussing ethical considerations
# 4. Providing actionable recommendations

if not df.empty:
    display_header("PHASE 4: INSIGHTS & RECOMMENDATIONS", level=2)

    # =================================================================
    # LINGUISTIC INSIGHTS
    # =================================================================
    if 'type' in df.columns and 'y' in locals():
        display_header("Linguistic Characteristics", level=3)

        # Create mask for fake news
        fake_mask = (y == 1)

        # Compare linguistic features
        insights = pd.DataFrame({
            'Metric': ['Word Count', 'Character Count', 'Sentiment Polarity'],
            'Fake News': [
                df[fake_mask]['word_count'].mean(),
                df[fake_mask]['char_count'].mean(),
                df[fake_mask]['sentiment'].mean()
            ],
            'Real News': [
                df[~fake_mask]['word_count'].mean(),
                df[~fake_mask]['char_count'].mean(),
                df[~fake_mask]['sentiment'].mean()
            ]
        })
        display("Feature Comparison:")
        display(insights)

        # Create 'is_fake' column for plotting if 'y' is available
        if 'y' in locals():
            df['is_fake'] = y.map({0: 'Real', 1: 'Fake'})

        # Visualize sentiment distribution
        plt.figure(figsize=(10, 6))
        sns.boxplot(
            x='is_fake',
            y='sentiment',
            data=df
        )
        plt.title('Sentiment Distribution by News Type')
        plt.xlabel('News Type')
        plt.ylabel('Sentiment Polarity')
        plt.show()

    # =================================================================
    # TEMPORAL PATTERNS
    # =================================================================
    if 'publish_hour' in df.columns:
        display_header("Temporal Publication Patterns", level=3)

        plt.figure(figsize=(12, 6))

        # If we have target information
        if 'y' in locals() and y.nunique() > 1:
            temp_df = df.copy()
            temp_df['is_fake'] = y.map({0: 'Real', 1: 'Fake'})

            # Group by hour and news type
            hourly_dist = temp_df.groupby(['publish_hour', 'is_fake']).size().unstack().fillna(0)

            # Create stacked bar plot
            hourly_dist.plot(kind='bar', stacked=True)
            plt.title('Publication Frequency by Hour of Day')
            plt.xlabel('Hour of Day')
            plt.ylabel('Number of Articles')
            plt.legend(title='News Type')
        else:
            # Simple count if no target available
            df['publish_hour'].value_counts().sort_index().plot(kind='bar')
            plt.title('Publication Frequency by Hour of Day')
            plt.xlabel('Hour of Day')
            plt.ylabel('Number of Articles')
        plt.tight_layout()
        plt.show()

    # =================================================================
    # DOMAIN ANALYSIS
    # =================================================================
    if 'domain' in df.columns:
        display_header("Domain Analysis", level=3)

        # Get top 10 domains by publication volume
        top_domains = df['domain'].value_counts().head(10)

        plt.figure(figsize=(10, 6))
        sns.barplot(x=top_domains.values, y=top_domains.index, palette='viridis')
        plt.title('Top 10 Domains by Publication Volume')
        plt.xlabel('Number of Articles')
        plt.tight_layout()
        plt.show()

    # =================================================================
    # ETHICAL CONSIDERATIONS
    # =================================================================
    display_header("Ethical Considerations", level=3)
    ethical_points = [
        "1. Subjectivity in labeling: 'Fake' vs 'Biased' requires careful consideration",
        "2. Model predictions should include confidence scores for transparency",
        "3. Potential for algorithmic bias against certain perspectives or sources",
        "4. Need for regular audits of model performance across different demographics"
    ]
    for point in ethical_points:
        display(point)

    # =================================================================
    # RECOMMENDATIONS
    # =================================================================
    display_header("Recommendations", level=3)

    # System improvements
    display("1. SYSTEM IMPROVEMENTS:")
    system_improvements = [
        "    - Implement model ensembles with deep learning architectures",
        "    - Integrate credibility scores from Media Bias/Fact Check",
        "    - Add social media engagement metrics as features",
        "    - Develop browser extension for real-time classification"
    ]
    for imp in system_improvements:
        display(imp)

    # Public education
    display("\n2. PUBLIC EDUCATION:")
    education_points = [
        "    - Create interactive media literacy modules for schools",
        "    - Launch public awareness campaigns about misinformation tactics",
        "    - Partner with libraries for community workshops"
    ]
    for point in education_points:
        display(point)

    # Policy advocacy
    display("\n3. POLICY ADVOCACY:")
    policy_points = [
        "    - Advocate for transparency in political advertising sources",
        "    - Support legislation against AI-generated disinformation",
        "    - Promote international standards for content verification"
    ]
    for point in policy_points:
        display(point)

    # =================================================================
    # PROJECT SUMMARY
    # =================================================================
    display_header("PROJECT SUMMARY", level=3)

    if 'f1' in locals() and 'roc_auc' in locals():
        display(f"Processed {len(df)} articles")
        display(f"Final model F1-score: {f1:.4f} (Target: 0.80)")
        display(f"ROC-AUC: {roc_auc:.4f}")

        # Check against target performance
        if f1 < 0.8:
            display_warning("⚠️ Model did not meet target F1-score of 0.80")
        else:
            display_success("✅ Model met target F1-score of 0.80")

    display(f"Project completed at {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
```
